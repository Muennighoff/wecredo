```
import tensorflow as tf
from transformers import TFAutoModelForSequenceClassification, TFDistilBertForSequenceClassification, DistilBertConfig

encoded_dataset_tf = dataset.map(preprocess_function_tf, batched=True)

def to_tf(ds, seq_len):
    """
    Creates a TF Dataset given basic tokenizer() encoding
    """
    # Turn into TF Tensors
    ds.set_format('tensorflow', columns=['input_ids', 'attention_mask', 'label'])
    # Extract TF Tensors
    features = {x: ds[x].to_tensor(default_value=0, shape=[None, seq_len]) for x in ['input_ids', 'attention_mask']}
    # Label is not a ragged tensor like the others & hence we separate it 
    tfds = tf.data.Dataset.from_tensor_slices((features, ds["label"]))
    return tfds

train_ds = to_tf(encoded_dataset_tf["train"], 512).shuffle(1000).batch(batch_size, drop_remainder=True)
val_ds = to_tf(encoded_dataset_tf["validation"], 512).batch(batch_size, drop_remainder=True)
```

```
class WrapperFinetuning(tf.keras.Model):
    def __init__(self, model, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.model = model
        # Track loss (Loss itself its CategoricalCrossEnt.)
        self.loss_tracker = tf.keras.metrics.Mean(name='loss') 

    @tf.function
    def train_step(self, data):
        # Data is a tuple of x, y for the CMNLI Finetuning data
        x, y = data
        with tf.GradientTape() as tape:
            # Forward pass
            y_pred = self.model(x, training=True).logits
            loss = self.compute_loss(y, y_pred)
            # Reduce loss to single digit
            loss = tf.reduce_mean(loss)
        # Compute gradients
        gradients = tape.gradient(loss, self.trainable_variables)
        # Update weights
        self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))
        # Update loss tracker
        self.loss_tracker.update_state(loss)  
        # Update metrics
        self.compiled_metrics.update_state(y, y_pred)

        return {m.name: m.result() for m in self.metrics}

    @tf.function
    def test_step(self, data):
        x, y = data
        # Compute predictions
        y_pred = self.model(x, training=False).logits
        loss = self.compute_loss(y, y_pred)
        # Updates the metrics tracking the loss
        self.loss_tracker.update_state(loss)  
        self.compiled_metrics.update_state(y, y_pred)

        return {m.name: m.result() for m in self.metrics}

    def compute_loss(self, labels, logits):
        if len(shape_list(logits)) == 1 or shape_list(logits)[1] == 1:
            loss_fn = tf.keras.losses.MeanSquaredError(reduction=tf.keras.losses.Reduction.NONE)
        else:
            # We use the below, as 3 labels
            loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(
                from_logits=True, reduction=tf.keras.losses.Reduction.NONE
            )

        return loss_fn(labels, logits)

def shape_list(tensor: tf.Tensor):
    """
    Deal with dynamic shape in tensorflow cleanly.
    Args:
        tensor (:obj:`tf.Tensor`): The tensor we want the shape of.
    Returns:
        :obj:`List[int]`: The shape of the tensor as a list.
    """
    dynamic = tf.shape(tensor)

    if tensor.shape == tf.TensorShape(None):
        return dynamic

    static = tensor.shape.as_list()

    return [dynamic[i] if s is None else s for i, s in enumerate(static)]
```

```
from datasets import load_dataset, load_metric
dataset = load_dataset("glue", task)
metric = load_metric('glue', task)

from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, use_fast=True)

# For fewshot / non-finetune performance we should add a prompt like "Is this positive or negative?"
def preprocess_function(examples):
    return tokenizer(examples["sentence"], truncation=True)

def preprocess_function_tf(examples):
    return tokenizer(examples["sentence"], truncation=True, padding="max_length", max_length=512)

# If Decoder is used:

label_map = {
    0: "negative </s>",
    1: "positive </s>",
    -1: "positive </s>", # Test Set is encoded as -1; Arbitrarily setting to positive here
}

def preprocess_targets(example):
    example["label"] = tokenizer(label_map[example["label"]], truncation=True)["input_ids"]
    return example

def preprocess_tf_targets(example):
    example["label"] = tokenizer(label_map[example["label"]], truncation=True, padding="max_length", max_length=2)["input_ids"]
    return example
```

```
from transformers import T5PreTrainedModel, T5EncoderModel 
from transformers.modeling_outputs import SequenceClassifierOutput

class T5ForSequenceClassification(T5PreTrainedModel):
    def __init__(self, config):
      super().__init__(config)
      self.encoder = T5EncoderModel(config)
      self.classifier = torch.nn.Linear(768, 2)
      self.loss_fct = torch.nn.CrossEntropyLoss()

      self.model_parallel = False

      self.init_weights()

    def forward(
        self,
        input_ids=None,
        attention_mask=None,
        head_mask=None,
        inputs_embeds=None,
        labels=None,
        output_attentions=None,
        output_hidden_states=None,
        return_dict=None,
    ):  

        encoder_outputs = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            inputs_embeds=inputs_embeds,
            head_mask=head_mask,
            output_attentions=output_attentions,
            output_hidden_states=output_hidden_states,
            return_dict=return_dict,
        )
        logits = self.classifier(encoder_outputs[0][:, 0])

        if labels is not None:
            loss_fct = torch.nn.CrossEntropyLoss()
            loss = loss_fct(logits.view(-1, 2), labels.view(-1))

        return SequenceClassifierOutput(
            loss=loss,
            logits=logits,
        )
```

```
class WrapperFinetuning(tf.keras.Model):
    def __init__(self, model, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.model = model
        self.classifier = tf.keras.layers.Dense(2)
        # Track loss (Loss itself its CategoricalCrossEnt.)
        self.loss_tracker = tf.keras.metrics.Mean(name='loss') 

    @tf.function
    def train_step(self, data):
        # Data is a tuple of x, y for the CMNLI Finetuning data
        x, y = data
        with tf.GradientTape() as tape:
            # Forward pass
            y_pred = self.model(x, training=True)[0]
            y_pred = self.classifier(y_pred[:, 0])
            loss = self.compute_loss(y, y_pred)
            # Reduce loss to single digit
            loss = tf.reduce_mean(loss)
        # Compute gradients
        gradients = tape.gradient(loss, self.trainable_variables)
        # Update weights
        self.optimizer.apply_gradients(zip(gradients, self.trainable_variables))
        # Update loss tracker
        self.loss_tracker.update_state(loss)  
        # Update metrics
        self.compiled_metrics.update_state(y, y_pred)

        return {m.name: m.result() for m in self.metrics}

    @tf.function
    def test_step(self, data):
        x, y = data
        # Compute predictions
        y_pred = self.model(x, training=False)[0]
        y_pred = self.classifier(y_pred[:, 0])
        loss = self.compute_loss(y, y_pred)
        # Updates the metrics tracking the loss
        self.loss_tracker.update_state(loss)  
        self.compiled_metrics.update_state(y, y_pred)

        return {m.name: m.result() for m in self.metrics}

    def compute_loss(self, labels, logits):
        if len(shape_list(logits)) == 1 or shape_list(logits)[1] == 1:
            loss_fn = tf.keras.losses.MeanSquaredError(reduction=tf.keras.losses.Reduction.NONE)
        else:
            # We use the below, as 3 labels
            loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(
                from_logits=True, reduction=tf.keras.losses.Reduction.NONE
            )

        return loss_fn(labels, logits)

def shape_list(tensor: tf.Tensor):
    """
    Deal with dynamic shape in tensorflow cleanly.
    Args:
        tensor (:obj:`tf.Tensor`): The tensor we want the shape of.
    Returns:
        :obj:`List[int]`: The shape of the tensor as a list.
    """
    dynamic = tf.shape(tensor)

    if tensor.shape == tf.TensorShape(None):
        return dynamic

    static = tensor.shape.as_list()

    return [dynamic[i] if s is None else s for i, s in enumerate(static)]
```
